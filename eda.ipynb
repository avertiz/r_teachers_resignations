```python
import math
import pandas as pd
import datetime
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data = pd.read_csv('teachers_analysis.csv').sample(n = 50000)
data = pd.read_csv('teachers_analysis.csv')
```

```python
grouped_data = (data[['post_month', 
                        'resignation_flair', 
                        'title_contains_resign', 
                        'text_contains_resign', 
                        'resignation_post', 
                        'negative_sentiment',
                        'neutral_sentiment',
                        'positive_sentiment']]
                    .groupby('post_month')
                    .agg(['sum','count','mean'])
                    .reset_index())
grouped_data['resignation_flair_pct'] = grouped_data['resignation_flair']['sum'] / grouped_data['resignation_flair']['count'] * 100
grouped_data['resignation_post_pct'] = grouped_data['resignation_post']['sum'] / grouped_data['resignation_post']['count'] * 100
# grouped_data
```

```python
# Resignation Flair Posts Plot 
x = grouped_data['post_month'].values
y1 = grouped_data['resignation_flair']['count'].values
y2 = grouped_data['resignation_flair']['sum'].values
fig = plt.figure(figsize = (24,12))
ax = plt.axes()
ax.margins(x=0)
ax.plot(x,y1,label="Post Count")
ax.plot(x,y2,label="Resignation Flair Count")
plt.legend(loc="upper left")
plt.grid()
ax.set_xticklabels(x, rotation = 45)
```

```python
# Resignation Posts Plot 
x = grouped_data['post_month'].values
y1 = grouped_data['resignation_post']['count'].values
y2 = grouped_data['resignation_post']['sum'].values
fig = plt.figure(figsize = (24,12))
ax = plt.axes()
ax.margins(x=0)
ax.plot(x,y1,label="Post Count")
ax.plot(x,y2,label="Resignation Post Count")
plt.legend(loc="upper left")
plt.grid()
ax.set_xticklabels(x, rotation = 45)
```

```python
# Resignation Flair Posts Pct Plot
x = grouped_data['post_month'].values
y1 = grouped_data['resignation_flair_pct'].values
fig = plt.figure(figsize = (24,12))
ax = plt.axes()
ax.margins(x=0)
ax.plot(x,y1,label="Pct of Posts Having Resignation Flair")
plt.legend(loc="upper left")
plt.grid()
ax.set_xticklabels(x, rotation = 45)
```

```python
# Resignation Posts Pct Plot
x = grouped_data['post_month'].values
y1 = grouped_data['resignation_post_pct'].values
fig = plt.figure(figsize = (24,12))
ax = plt.axes()
ax.margins(x=0)
ax.plot(x,y1,label="Pct of Posts Having Resignation In Them")
plt.legend(loc="upper left")
plt.grid()
ax.set_xticklabels(x, rotation = 45)
```

```python
# Sentiment Plot
x = grouped_data['post_month'].values
y1 = grouped_data['negative_sentiment']['mean'].values
y2 = grouped_data['neutral_sentiment']['mean'].values
y3 = grouped_data['positive_sentiment']['mean'].values
fig = plt.figure(figsize = (24,12))
ax = plt.axes()
ax.margins(x=0)
ax.plot(x,y1,label="Negative")
ax.plot(x,y2,label="Neutral")
ax.plot(x,y3,label="Positive")
plt.legend(loc="upper left")
plt.grid()
ax.set_xticklabels(x, rotation = 45)
```
